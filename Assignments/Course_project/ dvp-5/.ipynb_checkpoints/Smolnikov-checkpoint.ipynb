```python
feat_list = ['sex', 'age', 'suicides_no', 'population',
             '_gdp_for_year_($)_', 'gdp_per_capita_($)', 'generation', 'country','year'
            ]
y_list = ['suicides/100k_pop']
```

```python
feat_data = data[feat_list]
y_data = data[y_list]

amount_80 = int(len(data)*0.8)
train_data_X = feat_data[:amount_80]
train_data_y = y_data[:amount_80]

test_data_X = feat_data[amount_80:]
test_data_y = y_data[amount_80:]
```

```python
train_data_X.shape
```

```python
test_data_X.shape
```

```python
# Создаём модель леса из сотни деревьев
model = RandomForestRegressor(n_estimators=11,
                               criterion = 'mse',
                               bootstrap = True,
                               max_features = 'sqrt',
                               min_samples_split=42,
                               min_samples_leaf=2,
                               max_depth=37,
                               n_jobs=-1,
                               random_state=26,
                               verbose = 1,
                               oob_score = False,
                               warm_start = False,
#                                class_weight = None,
#                                max_samples = math.ceil(0.632*len(train_data_X))
                               )
# model = RandomForestClassifier(n_estimators=50,
#                                n_jobs=-1
#                                )

# Обучаем на тренировочных данных
model.fit(train_data_X, train_data_y)
```

```python

y_pred=model.predict(train_data_X)
```

```python
mean_absolute_error(train_data_y, y_pred)
```

```python
features = list(train_data_X.columns)
importances = model.feature_importances_
indices = np.argsort(importances)

top = 40
plt.title(f'Feature Importances top {top}')
plt.barh(range(len(indices))[:top], importances[indices][len(indices)-top:], color='b', align='center')
plt.yticks(range(len(indices))[:top], [features[i] for i in indices][len(indices)-top:])
plt.xlabel(f'Relative Importance top {top}')
plt.show()
```

год в данном наборе наблюдений имел неболшое значение, можно было рассматривать как задачу классификации

```python
pred_test_y=model.predict(test_data_X)
```

```python
import sklearn.metrics as metrics
print(metrics.r2_score(test_data_y, pred_test_y))
```

```python
mean_absolute_error(test_data_y, pred_test_y)
```

```python
fig, ax = plt.subplots()
ax.scatter(test_data_y, pred_test_y)
ax.plot([test_data_y.min(), test_data_y.max()], [test_data_y.min(), test_data_y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
model.score(test_data_X, test_data_y)
```

```python
mean_absolute_error(test_data_y, pred_test_y)
```
